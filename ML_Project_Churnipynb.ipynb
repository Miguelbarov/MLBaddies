```python
df.gender.replace({
    0: "Female",
    1: "Male"
}, inplace=True)

df.Partner.replace({
    0: "No",
    1: "Yes"
}, inplace=True)
df.Dependents.replace({
    0: "No",
    1: "Yes"
}, inplace=True)
df.PhoneService.replace({
    0: "No",
    1: "Yes"
}, inplace=True)
df.MultipleLines.replace({
    0: "No",
    1: "Yes",
    2: "No phone service"
}, inplace=True)
df.InternetService.replace({
    0: "No",
    1: "Fiber optic",
    2: "DSL"
}, inplace=True)
df.OnlineSecurity.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.OnlineBackup.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.DeviceProtection.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.TechSupport.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.StreamingTV.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.StreamingMovies.replace({
    0: "No",
    1: "Yes",
    2: "No internet service"
}, inplace=True)
df.Contract.replace({
    0: "Month-to-month",
    1: "One year",
    2: "Two year"
}, inplace=True)
df.PaperlessBilling.replace({
    0: "No",
    1: "Yes"
}, inplace=True)
df.PaymentMethod.replace({
    0: "Electronic check",
    1: "Mailed check",
    2: "Bank transfer (automatic)",
    3: "Credit card (automatic)"
}, inplace=True)

df.Churn.replace({
    0: "No",
    1: "Yes"
}, inplace=True)
```
